```python
def custom_binning_credit_score(credit_score):
    if credit_score < 580:
        return "Poor"
    elif credit_score < 670:
        return "Fair"
    elif credit_score < 740:
        return "Good"
    elif credit_score < 800:
        return "Very Good"
    elif credit_score <= 850:
        return "Excellent"
    else:
        assert True, "Credit Score out of range"
        
df["CreditScore_Binned"] = df["CreditScore"].apply(custom_binning_credit_score)
del df['CreditScore']   
df.head()
    
```

```python
df.to_csv("../data/processed/telco_data_binned.csv", index=False)
```
