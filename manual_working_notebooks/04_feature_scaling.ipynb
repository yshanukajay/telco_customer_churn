Importing Dependencies

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

Basic Processing

```python
df = pd.read_csv("../data/processed/telco_data_encoded.csv")
```

```python
df.head()
```

```python
df.Age.min(), df.Age.max()
```

```python
columns_to_scale = ['Age', 'Tenure', 'Balance', 'EstimatedSalary']

for col in columns_to_scale:
    scaler = StandardScaler()
    df[col] = scaler.fit_transform(df[[col]])
    
df.head()
```

```python
df.to_csv("../data/processed/telco_data_final.csv", index=False)
```
