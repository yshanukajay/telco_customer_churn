```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/telco_data_binned.csv")
```

Using Nominal Features

```python
nominal_variables = ['Geography', 'Gender'] 

geogrophical_dummies = pd.get_dummies(df['Geography'], prefix='Geography')
gender_dummies = pd.get_dummies(df['Gender'], prefix='Gender')

df_encoded = pd.concat([df, geogrophical_dummies], axis=1)
del df_encoded['Geography']

df_encoded = pd.concat([df_encoded, gender_dummies], axis=1)
del df_encoded['Gender']

df_encoded.sample(7)   
```

```python
encode_dict_credit_score = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}

df_encoded['CreditScore_Binned'] = df_encoded['CreditScore_Binned'].map(encode_dict_credit_score)

df_encoded.head()
```

```python
df_encoded.to_csv("../data/processed/telco_data_encoded.csv", index=False)
```
